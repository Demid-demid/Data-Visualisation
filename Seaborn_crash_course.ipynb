```python
import seaborn as sb
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 


%matplotlib inline
```

```python
sb.set(style="darkgrid")#Set aesthetic parameters in one step.
tips = sb.load_dataset("tips")
tips.head()
#or 
#cars = pd.read_csv('/Users/demiddemidov/Desktop/fuel_econ.csv')
#print(cars.head())
```

```python
tips["smoker"].value_counts()# will shom me the index Smoker and count the relation ship
```

```python
sb.regplot(x="total_bill", y="tip", data=tips,)#Plot data and a linear regression model fit.
#dir(sb.FacetGrid)#the things that we can apply on Faset Grid
```

```python
sb.relplot(x="total_bill", y="tip", data=tips,hue="smoker", style="time")# reL plot to see ralation between smoker and non
#hue we can  use for the "sex", "time" and etc.
```

```python
#b.relplot(x="total_bill", y="tip", data=tips, hue="size")
sb.relplot(x="total_bill", y="tip", data=tips, hue="size", palette="RdBu")
# hue = means the color changing
# size with color< the more the darker  
#palette to change color < check  sb.color_palette to  all  functions 
```

```python
sb.set_palette("Greens_r") # to  change color  use set  palette
sb.relplot(x="total_bill", y="tip", data=tips, size="size" )
#pay attention to size = "size" < the bigger data< the bigger size of circle 

```

```python
sb.set_palette("winter_r") # to  change color  use set  palette
sb.relplot(x="total_bill", y="tip", data=tips, size="size", sizes = (20,500))
#sizes will change the actual size of circles 
```
