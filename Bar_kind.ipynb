```python
import pandas as pd
import seaborn as sb
import numpy as np 
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
titanic = sb.load_dataset("titanic")
titanic.head()
```

```python
sb.catplot(x="sex",y="survived",kind= "bar", hue="class", data=titanic)
```

```python
sb.catplot(x="deck",kind= "count", palette="ch:0.55", data=titanic)
#We can see that most of the people were traveling in the C deck
#we used kind= "count" for that 
```
